O eixo Y vai de 1 até 4. E quando o eixo X não é específicado, o gráfico apresentará de 0 até o número de amostras do Y menos 1.

```python
x = np.arange(0, 10, 0.2)
y = np.sin(x)
```

```python
plt.plot(x, y)

plt.title('Seno')
plt.xlabel('X')
plt.ylabel('Y / sen(x)')
plt.grid(True)
```

# Plotagem básica II & subplots

1. Plotando mais de um gráfico em um mesmo axes
1. Definindo legendas
1. Definindo os valores contidos nos eixos
1. Subplots

```python
# Para que os gráficos fiquem numa linha no notebook
%matplotlib inline
```

```python
x = np.arange(0, 10, 0.01)
y1 = np.sin(x) # primeira curva
y2 = np.cos(x) # segunda curva
y3 = np.sin(x) + np.cos(3 * x)
```

```python
# plt.plot(x, y1, x, y2)
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x) + cos(3x)')

plt.xticks(range(11))

plt.legend()

# Posicionar no canto superior direito
# plt.legend(loc='upper right')
```

## Subplots

```python
plt.figure(figsize=(12, 4))

# 1p: número de linhas
# 2p: número de colunas
# 3p: plot (em qual coluna o gráfico vai ficar)
plt.subplot(1, 3, 1)
plt.plot(x, y1)

plt.subplot(1, 3, 2)
plt.plot(x, y2)

plt.subplot(1, 3, 3)
plt.plot(x, y3)

# Otimizar os espaçamentos entre os axes
plt.tight_layout()
```

```python
y4 = np.cos(x) + np.sin(3 * x)

plt.subplot(2, 2, 1)
plt.plot(x, y1)

plt.subplot(2, 2, 2)
plt.plot(x, y2)

plt.subplot(2, 2, 3)
plt.plot(x, y3)

plt.subplot(2, 2, 4)
plt.plot(x, y4)

plt.tight_layout()
```

```python
df_eua = df[df['Country'] == 'EUA']
```

```python
df_eua
```

#### `subplots`

```python
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(14,5))

axes[0].plot(df_brasil['Year'], df_brasil['Life Expectancy'], marker='o')
axes[0].set_xlabel('Ano')
axes[0].set_ylabel('Expectativa de vida')
axes[0].set_title('Dados do Brasil')

axes[1].plot(df_eua['Year'], df_eua['Life Expectancy'], marker='o')
axes[1].set_xlabel('Ano')
axes[1].set_ylabel('Expectativa de vida')
axes[1].set_title('Dados dos Estados Unidos')
```

```python
figure, axes = plt.subplots(nrows=2, ncols=2, figsize=(14,9))

axes[0][0].plot(df_brasil['Year'], df_brasil['Life Expectancy'], marker='o')
axes[0][0].set_xlabel('Ano')
axes[0][0].set_ylabel('Expectativa de vida')
axes[0][0].set_title('Dados do Brasil')

axes[0][1].plot(df_eua['Year'], df_eua['Life Expectancy'], marker='o')
axes[0][1].set_xlabel('Ano')
axes[0][1].set_ylabel('Expectativa de vida')
axes[0][1].set_title('Dados dos Estados Unidos')
```
