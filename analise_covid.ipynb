# Caso confirmados de COVID 19

```python
import pandas as pd
from urllib.request import urlretrieve
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

urlretrieve(url, 'datasets/global_cases_covid19.csv')
```

```python
df_covid = pd.read_csv('datasets/global_cases_covid19.csv')
```

```python
df_covid.drop(['Lat', 'Long'], axis=1, inplace=True)
```

```python
df_country = df_covid.groupby('Country/Region').sum()
```

```python
df_country
```

```python
df_country.loc['Afghanistan']
```

```python
date = df_country.loc['Brazil'].index
cases = df_country.loc['Brazil'].values
```

```python
s_brazil = df_country.loc['Brazil']
```

```python
# 30 últimos dias
tamanho = len(s_brazil)

plt.figure(figsize=(14, 8))
plt.xticks(rotation=60)
plt.bar(s_brazil.index[tamanho - 30: tamanho], s_brazil.values[tamanho - 30: tamanho])
plt.title('Total de casos confirmados de COVID-19 no Brasil')
```
